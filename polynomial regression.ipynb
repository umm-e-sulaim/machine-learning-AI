```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

# Generate sample data
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, (100, 1))

# Create polynomial regression model (degree 2)
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Fit model
polyreg.fit(X, y.ravel())

# Predict
y_pred = polyreg.predict(X)

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Polynomial fit (degree 2)')
plt.title('Polynomial Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Save plot
plt.savefig('polynomial_regression.png')
```

```python
# prompt: test the model if the output is not correct retrain the save model again

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import joblib
from sklearn.metrics import mean_squared_error

# Generate sample data
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, (100, 1))

# Create polynomial regression model (degree 2)
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())

# Fit model
polyreg.fit(X, y.ravel())

# Predict
y_pred = polyreg.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")

# Define acceptable error threshold
threshold = 0.5

# Retrain if error exceeds threshold
if mse > threshold:
    print("Model performance is below the threshold. Retraining...")
    # Add more data or change model parameters if necessary
    # For this example, just refit the model
    polyreg.fit(X, y.ravel())
    y_pred = polyreg.predict(X)
    mse = mean_squared_error(y, y_pred)
    print(f"Retrained Model Mean Squared Error: {mse}")

# Save the model
joblib.dump(polyreg, 'polynomial_regression_model.pkl')

# Load and test the saved model
loaded_model = joblib.load('polynomial_regression_model.pkl')
y_pred_loaded = loaded_model.predict(X)
mse_loaded = mean_squared_error(y, y_pred_loaded)
print(f"Loaded Model Mean Squared Error: {mse_loaded}")

# Plot results
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred_loaded, color='red', label='Polynomial fit (degree 2)')
plt.title('Polynomial Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.savefig('polynomial_regression.png')
plt.show()
```

```python
# prompt: test the model if the output is not correct retrain the save model again

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import joblib
from sklearn.metrics import mean_squared_error, r2_score

# Load the saved model (if it exists)
try:
    loaded_model = joblib.load('polynomial_regression_model.pkl')
    print("Loaded existing model.")
except FileNotFoundError:
    print("Model file not found. Training a new model.")
    loaded_model = None

# Generate sample data
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, (100, 1))


# Train a new model or retrain the existing one if necessary
if loaded_model is None:
    # Create and train a new polynomial regression model
    degree = 2
    polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    polyreg.fit(X, y.ravel())
    loaded_model = polyreg
else:
    # Evaluate the loaded model
    y_pred_loaded = loaded_model.predict(X)
    mse_loaded = mean_squared_error(y, y_pred_loaded)
    print(f"Loaded Model Mean Squared Error: {mse_loaded}")

    # Retrain if MSE is above the threshold
    threshold = 0.5
    if mse_loaded > threshold:
      print("Retraining the model...")
      degree = 2
      polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
      polyreg.fit(X, y.ravel())
      loaded_model = polyreg
      y_pred_loaded = loaded_model.predict(X)
      mse_loaded = mean_squared_error(y, y_pred_loaded)
      print(f"Retrained Model Mean Squared Error: {mse_loaded}")


# Save the (potentially retrained) model
joblib.dump(loaded_model, 'polynomial_regression_model.pkl')

# Make predictions using the loaded model
y_pred = loaded_model.predict(X)


# Evaluate the final model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Final Model Mean Squared Error: {mse}")
print(f"Final Model R-squared: {r2}")

#Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', label='Polynomial fit (degree 2)')
plt.title('Polynomial Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.savefig('polynomial_regression.png')
plt.show()
```

```python
# prompt: compare both of the save mode and tell which model is best

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
import joblib
from sklearn.metrics import mean_squared_error, r2_score

# Function to train and evaluate a model
def train_and_evaluate_model(X, y, degree=2, threshold=0.5, filename='polynomial_regression_model.pkl'):
    """Trains, evaluates, and saves a polynomial regression model.

    Args:
        X: Input features.
        y: Target variable.
        degree: Degree of the polynomial.
        threshold: MSE threshold for retraining.
        filename: Filename to save the model.

    Returns:
        A tuple containing the trained model and its MSE.
    """
    polyreg = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    polyreg.fit(X, y.ravel())
    y_pred = polyreg.predict(X)
    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    print(f"Initial Model Mean Squared Error: {mse}")
    print(f"Initial Model R-squared: {r2}")

    if mse > threshold:
        print("Model performance is below the threshold. Retraining...")
        polyreg.fit(X, y.ravel())
        y_pred = polyreg.predict(X)
        mse = mean_squared_error(y, y_pred)
        r2 = r2_score(y, y_pred)
        print(f"Retrained Model Mean Squared Error: {mse}")
        print(f"Retrained Model R-squared: {r2}")

    joblib.dump(polyreg, filename)
    return polyreg, mse


# Generate sample data
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, (100, 1))


# Train and evaluate the initial model
model, initial_mse = train_and_evaluate_model(X,y)

# Load and test the saved model
loaded_model = joblib.load('polynomial_regression_model.pkl')
y_pred_loaded = loaded_model.predict(X)
mse_loaded = mean_squared_error(y, y_pred_loaded)
print(f"Loaded Model Mean Squared Error: {mse_loaded}")


# Generate new data for prediction (or use your actual test data)
X_test = np.linspace(-3, 3, 50).reshape(-1, 1)  # Example test data
y_test = 0.5 * X_test**2 + X_test + 2 + np.random.normal(0, 1, (50, 1))

# Make predictions using the loaded model
y_pred_test = loaded_model.predict(X_test)

# Evaluate the model on the new data
mse_test = mean_squared_error(y_test, y_pred_test)
r2_test = r2_score(y_test, y_pred_test)

print(f"Test Mean Squared Error: {mse_test}")
print(f"Test R-squared: {r2_test}")


#Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred_loaded, color='red', label='Polynomial fit (degree 2)')
plt.title('Polynomial Regression Example')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.savefig('polynomial_regression.png')
plt.show()
```
